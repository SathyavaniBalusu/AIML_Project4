```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from PIL import Image
from array import array


im = np.array(Image.open('Car.jpg'))
print(im.shape, im.dtype)
```

```python
plt.imshow(im)
```

```python
img = np.array(im).reshape((596, 900*3))
print(img.shape)
```

```python
pca = PCA(n_components=32).fit(img) 
img_transformed = pca.transform(img) 
print(img_transformed.shape)
print(np.sum(pca.explained_variance_ratio_) )
```

```python
image_recon = pca.inverse_transform(pca.fit_transform(img))
plt.figure(figsize=[20,15])
plt.imshow(image_recon)
print(image_recon.shape)
```

The possible dimensionality reduction techniques are:
    1. PCA- PCA combines the features and reduce the number of dimentions. This technique will improve the accuracy score.
    2. Ridge- This is used to analyse any data that suffers from multicollinearity. It shrinks the parameters
    to prevent multicollinearity.
    3. Lasso-This method shriks the non important features to zero thereby reducing the no of dimentions.
